```python
import pandas as pd
import numpy as np
```

```python
def isfloat(value):
  try:
    float(value)
    return True
  except ValueError:
    return False
```

```python
df = pd.read_csv('dataset03.csv')
print(df.columns)

df = df.dropna(axis=1)
print(df.columns)

x = df.values[:,:-1]
y = df.values[:,-1]
```

```python
m,n = np.shape(x)
for i in range(m):
    for j in range(n):
        if isfloat(x[i,j]):
            x[i,j] = float(x[i,j])
        else:
#             print(x[i,j])
            x[i,j] = 0.0
```

```python
x_filled = x.copy()
for i in range(n):
    mean = np.mean(x[np.nonzero(x[:,i])])
    sigma = np.sqrt(np.var(x[np.nonzero(x[:,i])]))
#     print(mean,sigma)
    for j in range(m):
        if x[j,i]==0:
            x_filled[j,i] = np.random.normal(loc=mean,scale=sigma)
# print(x_filled)
```

```python
for i in range(m):
    if y[i]=='B':
        y[i] = 0
    else:
        y[i]=1
```

```python
import matplotlib.pyplot as plt
mal = np.nonzero(y)
mal = mal[0]
benign = [i for i in range(len(y)) if i not in mal]

for i in range(n):
    plt.scatter(mal,x[mal,i],color='r')
#     plt.show()
    
    plt.scatter(benign,x[benign,i],color='b')
    mean = np.mean(x[:,i])
    std = np.sqrt(np.var(x[:,i]))
    plt.axis([0, 600,mean-2.5*std, mean+2.5*std])
    plt.ylabel('feature value')
    plt.xlabel('record num')
    plt.title('Feature '+str(i)+' : '+df.columns[i])
    plt.show()
```

```python
# Descritizing
for i in range(n):
    mean = np.mean(x[:,i])
    subtracted = x[:,i] - mean* np.ones(np.shape(x[:,i]))
    for j in range(m):
        if subtracted[j]>0:
            x[j,i] = 1
        else:
            x[j,i] = 0
#     print(x[:,i])
```

```python
from sklearn.model_selection import KFold
from sklearn.naive_bayes import MultinomialNB
def ten_fold_cv(xx,yy):
    kf = KFold(n_splits=10,shuffle=True)
    i = 0
    for train_index, test_index in kf.split(xx):
        i += 1
        X_train, X_test = xx[train_index], xx[test_index]
        y_train, y_test = yy[train_index], yy[test_index]
        clf = MultinomialNB()
        clf.fit(X_train, np.array(y_train,dtype=np.int32))
        pred = clf.predict(X_test)
        print('fold',i,', accuarcy',np.mean(np.abs(y_test-pred)))
ten_fold_cv(x,y)
```

```python
# using feature 0,1,6
ten_fold_cv(x[:,[0,1,6]],y)
```

```python
# using feature 8,9,5
ten_fold_cv(x[:,[8,9,10]],y)
```

```python
# using feature 13,12
ten_fold_cv(x[:,[13,12]],y)
```
