```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset01.csv')
```

```python
x = df.values[:,:-1]
y = df.values[:,-1]
x_train = x[:500,:]
x_test = x[500:,:]
y_train = y[:500]
y_test = y[500:]
```

```python
print(np.shape(x))
print(np.shape(y))
```

```python
# Plot
for i in range(8):
    plt.scatter(x[:,i], y)
    plt.title('Scatter plot feature '+str(i))
    plt.xlabel('x['+str(i)+']')
    plt.ylabel('y')
    plt.show()
```

```python
def abline(slope, intercept):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--')
    plt.show()
```

```python
# Simple Linear Regression
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error,r2_score
#https://www.statsmodels.org/stable/generated/statsmodels.regression.linear_model.RegressionResults.html
for i in range(8):
    print('------------------------- feature:',str(i),'-------------------------')
    model = sm.OLS(y_train,x_train[:,i]).fit()
    pred = model.predict(x_test[:,i]) # make the predictions by the model
    pred_train = model.predict(x_train[:,i])
    plt.scatter(x_train[:,i], y_train)
    plt.xlabel('x['+str(i)+']')
    plt.ylabel('y')
    abline(model.params[0],0)
    print('params',model.params)
    print('R2(test)',r2_score(y_test,pred))
    print('RSS(test)',len(y_test) * mean_squared_error(y_test,pred))
    print('R2(train)',r2_score(y_train,pred_train))
    print('RSS(train)',len(y_train) * mean_squared_error(y_train,pred_train))
    print('Sigma Squared',len(y_train) * mean_squared_error(y_train,pred_train)/(len(y_train)-2))
    print('std for params',model.bse)
```

```python
# Linear Regression
import statsmodels.api as sm
#https://www.statsmodels.org/stable/generated/statsmodels.regression.linear_model.RegressionResults.html
for i in range(8):
    print('------------------------- feature:',str(i),'-------------------------')
    X = sm.add_constant(x_train[:,i])
    model = sm.OLS(y_train,X).fit()
    x_t = sm.add_constant(x_test[:,i])
    pred = model.predict(x_t) # make the predictions by the model
    pred_train = model.predict(X)
    plt.scatter(x_train[:,i], y_train)
    plt.xlabel('x['+str(i)+']')
    plt.ylabel('y')
    abline(model.params[0],0)
    print('params',model.params)
    print('R2(test)',r2_score(y_test,pred))
    print('RSS(test)',mean_squared_error(y_test,pred))
    print('R2(train)',r2_score(y_train,pred_train))
    print('RSS(train)',mean_squared_error(y_train,pred_train))
    print('Sigma Squared',len(y_train) * mean_squared_error(y_train,pred_train)/(len(y_train)-2))
    print('std for params',model.bse)
```

```python
# Leave one out: Training n models

from sklearn.model_selection import LeaveOneOut
loo = LeaveOneOut()
RSSs = []
for train_index, test_index in loo.split(X):
    X_train, X_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]

    X = sm.add_constant(X_train)
    model = sm.OLS(y_train,X).fit()
    x_t = sm.add_constant(X_test)
    RSSs.append(model.ssr)
print(np.mean(RSSs))
```

```python
# Leave one out: Training 1 model

def leave_one_out(xx,yy):

    leverage = xx.dot(np.linalg.inv(xx.T.dot(xx))).dot(xx.T)

    X = sm.add_constant(xx)
    model = sm.OLS(yy,xx).fit()
    pred = model.predict(xx)
    r = [((yy[i]-pred[i])*(yy[i]-pred[i]))/((1-leverage[i,i])*(1-leverage[i,i])) for i in range(len(yy))]
    return np.sum(r)
r =leave_one_out(x,y)
print(r)
# print(model.ssr)

```

```python
# PART vav: backward model selection


import statsmodels.api as sm
RSSs = []
cvs = []
feature_ids = [0,1,2,3,4,5,6,7]

for i in feature_ids:
    print('------------------------- feature:', str(i) ,' removed -------------------------')
    fset = feature_ids.copy()
    fset.remove(i)
    X = sm.add_constant(x[:,fset])
    model = sm.OLS(y,X).fit()
    x_t = sm.add_constant(x[:,fset])
    predictions = model.predict(x_t) # make the predictions by the model
    r = leave_one_out(x[:,fset],y)
    RSSs.append(model.ssr)
    cvs.append(r)
    print(r)
chosen = np.argmin(cvs)
print('Feature',feature_ids[chosen],'is choosed to be removed')
RSS_during_removing.append(RSSs[chosen])
```

```python
# PART Ze: different ratios for train ans test
RSSs = []
train_RSSs = []
fset = [0,2,3,6]
ratios = np.arange(0.05, 1.0, 0.025)
for ratio in ratios:
    train_size = int(len(x) * ratio)
    x_train = x[:train_size,:]
    x_test = x[train_size:,:]
    y_train = y[:train_size]
    y_test = y[train_size:]
    X = sm.add_constant(x_train[:,fset])
    model = sm.OLS(y_train,X).fit()
    x_t = sm.add_constant(x_test[:,fset])
    predictions = model.predict(x_t) 
    predictions2 = model.predict(X) 
    RSSs.append(len(y_test)*mean_squared_error(y_test,predictions))
    train_RSSs.append(len(y_train)*mean_squared_error(y_train,predictions2))
plt.plot(ratios,RSSs)
plt.xlabel('Train_size/all')
plt.ylabel('Test RSS')
plt.show()
plt.plot(ratios,train_RSSs)
plt.xlabel('Train_size/all')
plt.ylabel('Train RSS')
plt.show()
```
