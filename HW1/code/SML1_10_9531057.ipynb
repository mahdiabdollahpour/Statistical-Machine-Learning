```python
# Question 10
import random
import matplotlib.pyplot as plt
import math
import numpy as np

n = 100000
res1 = []
res2 = []
for i in range(n):
    x1 = np.random.binomial(100,0.06)
    x2 = np.random.poisson(6)
    res1.append(x1)
    res2.append(x2)
plt.hist(res1,bins=range(min(res1), max(res1), 1))
plt.show()
plt.hist(res2,bins=range(min(res2), max(res2), 1))
plt.show()
print("Binomial mean",np.mean(res1))
print("Poisson mean",np.mean(res2))
```
