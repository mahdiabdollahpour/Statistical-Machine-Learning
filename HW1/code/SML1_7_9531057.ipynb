```python
# Question 7

import random
import matplotlib.pyplot as plt
import math
import numpy as np

nums = [1,100,1000,10000]
n = 10000
epsilon = 0.002
a = [0,0,0,0]
idx = -1
for i in nums:
    idx+=1
    rs = []
    for j in range(n):
        r = np.random.normal(0,1/i)
        rs.append(r)
        if abs(r) > epsilon:
            a[idx]+=1
    plt.hist(rs, normed=True, cumulative=True, label='CDF',
         histtype='step', alpha=0.8, color='k')
    plt.show()
print(a[0]/n)
print(a[1]/n)
print(a[2]/n)
print(a[3]/n)
    
```
