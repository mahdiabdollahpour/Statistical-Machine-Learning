```python
import random
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
# Question 1
# A
p = 0.3
n = 1000
res = []
proportions = []
for i in range (n):
    r = random.random()
    if r<=p:
        res.append(1)
        
    else:
        res.append(0)
    proportions.append(np.mean(res))
plt.scatter(range(n),proportions)
plt.show()
```
