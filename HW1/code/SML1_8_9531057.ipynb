```python
# Question 8

import random
import matplotlib.pyplot as plt
import math
import numpy as np

n = 1000
a = []
b = []
for i in range(n):
    x = np.random.uniform(0,1)
    y = np.random.uniform(0,1)
    a.append(x-y)
    b.append(x/y)
    

hist, bins = np.histogram(a, bins=100)
bin_centers = (bins[1:]+bins[:-1])*0.5
plt.plot(bin_centers, hist)
plt.show()

plt.hist(b,range(-2,100))
plt.show()
```
