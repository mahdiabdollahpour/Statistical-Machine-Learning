```python
# Question 9 

import random
import matplotlib.pyplot as plt
import math
import numpy as np

# Normal
n = 10000
rs = []
means = []
for i in range(n):
    r = np.random.normal(0,1)
    rs.append(r)
    means.append(np.mean(rs))
plt.scatter(range(n),means)
plt.show()

# Cauchy
rs = []
means = []
for i in range(n):
    r = np.random.standard_cauchy()
    rs.append(r)
    means.append(np.mean(rs))
plt.scatter(range(n),means)
plt.show()
```
